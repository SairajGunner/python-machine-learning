```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
```

```python
iris.target_names
```

```python
clf_dt = DecisionTreeClassifier()
```

```python
clf_dt.fit(iris.data, iris.target)
```

```python
iris.feature_names
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
plt.figure(figsize=(12, 12))
tree.plot_tree(clf_dt, fontsize=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_rf = RandomForestClassifier(n_estimators=5, max_depth=3)
```

```python
clf_rf.fit(iris.data, iris.target)
```

```python
clf_rf.estimators_
```

```python
for estimator in clf_rf.estimators_:
    plt.figure(figsize=(12, 12))
    tree.plot_tree(estimator, fontsize=10)
```
